```python
#plotting 2 lists
#there are 5 subjects psychology,nano technology,medicine,oncology,anthropology two students have applied to do masters in any of these 5 feilds with the line plot show in which feild they can excel the cutoff for every subject is 70
student1=[67,45,73,71,57]
student2=[43,67,68,53,85]
subject=["psychology ","nano technology","medicine","oncology","anthroplogy"]
cutoff=[70,70,70,70,70]
#plt.plot(lebels,legend, colour, width, markers)
plt.scatter(subject,student1,label="student1",color="pink",marker="p")
plt.scatter(subject,student2,label="student2",color="blue",marker="H")
plt.plot(subject,cutoff,label="cutoff",linestyle=":",linewidth=3)
plt.xlabel("subjects")
plt.ylabel("marks")
plt.legend()
```

```python
#an India vs Australia match is on live,India has given 298 as target for australia,the runs for every five overs is given as follows
#india=[25,51,84,131,160,189,220,250,267,297],australia has completed just 25 overs australia=[15,41,94,110,151],plot overs vs runs for both the countries with india as a blue line and australia as yellow line,give appropriate labels and legand,the marker for india is flag,aus is plus,take overs from arange function,create runs as arrays
#marker=">"==flag
india=np.array([25,51,84,131,160,189,220,250,267,297])
overs1=np.arange(5,51,5)
australia=np.array([15,41,94,110,151])
overs2=np.arange(5,26,5)
plt.plot(overs1,india,color="blue",label="india",marker=">",linewidth=2)
plt.plot(overs2,australia,color="yellow",label="australia",marker="+",linewidth=2)
plt.legend()
plt.title("IND vs AUS")
plt.xlabel("overs")
plt.ylabel("runs")
plt.grid()
```
