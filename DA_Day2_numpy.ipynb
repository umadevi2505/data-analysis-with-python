```python
#vstack-prints matrices one on other
A=np.array([12,13,14,15]).reshape(2,2)
B=np.array([23,56,78,10]).reshape(2,2)
print(A)
print("----------")
print(B)
print("after vertically stacking the arrays")
print(np.vstack((A,B)))
```

```python
#hstack-prints matrices one side other
A=np.array([12,13,14,15]).reshape(2,2)
B=np.array([23,56,78,10]).reshape(2,2)
print(A)
print("----------")
print(B)
print("after horizontally stacking the arrays")
print(np.hstack((A,B)))
```

```python
A=np.array([12,13,14,15]).reshape(2,2)
B=np.array([23,56,78,10]).reshape(2,2)
print(A)
print("----------")
print(B)
print("----------")
print(np.stack((A,B),axis=0))
```
