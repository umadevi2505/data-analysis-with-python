```python
#Hypothesis testing:
#import
import numpy as np
from scipy import stats
```

```python
#two grps of data
data1=np.array([28,21,26,29,23])
data2=np.array([21,27,25,28,19])
print(data1.mean())
print(data2.mean())
```

```python
#perform t-test assuming unequal variances
t_stat,p_value=stats.ttest_ind(data1,data2,equal_var=False)
```

```python
#PRINT RESULTS
print("T-statistic:",t_stat)
print("p-value",p_value)
```

```python
alpha=0.05
```

```python
if p_value < alpha:
  print("reject the null hypothesis.the means are significant:")
else:
  print("fail to reject the null hypothesis. the means are not significant")
```

REGRESSION ANALYSIS

```python
import pandas as pd
df=pd.read_csv("/content/Vaccine.csv")
df
```

```python
a=df['Efficiency'].mean()
print("Mean Efficiency",a)
b=df['Rec_Rate'].mean()
print("Mean REcovery",b)
group_A = df['Efficiency']
group_B = df['Rec_Rate']
t_stat,p_value = stats.ttest_ind(group_A,group_B)
print("t_val",t_stat)
print("p_val",p_value)
```

```python
alpha=0.05
```

```python
if p_value>alpha:
  print("H0-related")
else:
  print("H1-not related")
```

ANOVA (ONE WAY ANOVA)

```python
dfA=(df[df['Vaccine']=='A']['Efficiency'])
dfB=(df[df['Vaccine']=='B']['Efficiency'])
dfA
```

```python
f_stat,p_value=stats.f_oneway(dfA,dfB)
print("F-statistic:",f_stat)
print("p-value:",p_value)
```

```python
if p_value>alpha:
  print("H0-related")
else:
  print("H1-not related")
```
